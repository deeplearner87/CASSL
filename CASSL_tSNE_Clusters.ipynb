```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import preprocessing
```

**For E-GEOD-81547 dataset**

```python
missing = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Semi-Supervised_Learning/Results_New/81547_missingAnnotations.csv", header=0, index_col=None, sep=",")
predicted = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Semi-Supervised_Learning/Results_New/81547_predictedAnnotations.csv", header=0, index_col=None, sep=",")
```

```python
x_embedded1 = missing[['x','y']]
x_embedded2 = predicted[['x','y']]
```

```python
label1 = missing[['label']]
label2 = predicted[['label']]
```

```python
label_encoder1 = preprocessing.LabelEncoder()
label_encoder2 = preprocessing.LabelEncoder()  
# Encode labels in column 'species'.
missing['class'] = label_encoder1.fit_transform(np.ravel(label1))
predicted['class'] = label_encoder2.fit_transform(np.ravel(label2))
```

```python
l1 = missing[['class']]
l2 = predicted[['class']]
```

```python
cols = ['green', 'grey', 'blue', 'cyan', 'purple', 'magenta', 'red']
```

```python
x_embedded1 = np.array(x_embedded1)
x_embedded2 = np.array(x_embedded2)
```

Plotting tSNE Clusters

```python
# Visualizing original data 
fig, ax = plt.subplots(figsize=(18, 10))
fig.suptitle('Clusters with missing annotations', fontsize=20)
ax.set_xlabel('tSNE_1', fontsize=18)
ax.set_ylabel('tSNE_2', fontsize=18)
for i in np.unique(l1):
  if i==6:
    plt.scatter(x_embedded1[l1['class']==i, 0], x_embedded1[l1['class']==i, 1], marker="x", s = 30, c = cols[i], cmap = 'rainbow')  
  else:
    plt.scatter(x_embedded1[l1['class']==i, 0], x_embedded1[l1['class']==i, 1], marker="o", s = 20, c = cols[i], cmap = 'rainbow')
plt.show()
```

```python
# Visualizing predicted data 
fig, ax = plt.subplots(figsize=(18, 10))
fig.suptitle('Clusters with predicted annotations', fontsize=20)
ax.set_xlabel('tSNE_1', fontsize=18)
ax.set_ylabel('tSNE_2', fontsize=18)
for i in np.unique(l2):
  if i==6:
    plt.scatter(x_embedded2[l2['class']==i, 0], x_embedded2[l2['class']==i, 1], marker="x", s = 30, c = cols[i], cmap = 'rainbow')  
  else:
    plt.scatter(x_embedded2[l2['class']==i, 0], x_embedded2[l2['class']==i, 1], marker="o", s = 20, c = cols[i], cmap = 'rainbow')
plt.show()
```
